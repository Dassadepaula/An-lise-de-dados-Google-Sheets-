```python
import pandas as pd
import plotly.express as px
```

# #Importando as planilhas do Google Sheet


```python
from os import name
df_Principal = pd.read_excel ("/content/Acoes_puras.xlsx", sheet_name="Principal")
df_Principal
```

```python
df_CHATGPT = pd.read_excel ("/content/Acoes_puras.xlsx", sheet_name="CHATGPT")
df_CHATGPT
```

```python
df_Total_de_acoes = pd.read_excel ("/content/Acoes_puras.xlsx", sheet_name="Total_de_acoes")
df_Total_de_acoes
```

```python
df_Ticker = pd.read_excel ("/content/Acoes_puras.xlsx", sheet_name="Ticker")
df_Ticker
```

Mudança nos nomes das colunas


```python
df_Principal = df_Principal[['Ativo',	'Data',	'Último (R$)',	'Var. Dia (%)']].copy()
df_Principal
```

Adicionando as demais colunas da planilha

```python
df_Principal = df_Principal.rename(columns={'Último (R$)':'valor_final','Var. Dia (%)':'var_dia_pct'}).copy()
df_Principal
```

```python
df_Principal['var_pct'] = df_Principal['var_dia_pct'] / 100
df_Principal['valor_inicial'] = df_Principal['valor_final'] / (df_Principal['var_pct'] + 1)
df_Principal
```

```python
df_Principal = df_Principal.merge(df_Total_de_acoes,left_on='Ativo',right_on='Código',how='left')
df_Principal
```

Apagando uma coluna

```python
df_Principal.drop(columns=['Código'])
```

```python
df_Principal['var_real'] = (df_Principal['valor_final'] - df_Principal['valor_inicial']) * df_Principal['Qtde. Teórica']
df_Principal
```

Formatação numérica

```python
pd.options.display.float_format = '{:.2f}'.format
df_Principal
df_Principal['Qtde. Teórica'] = df_Principal['Qtde. Teórica'].astype(int)
df_Principal
```

Renomeando a coluna 'Qtde.Teórica'

```python
df_Principal = df_Principal.rename(columns={'Qtde. Teórica':'Qtd_teorica'}).copy()
df_Principal
```

```python
df_Principal['Resultado'] = df_Principal['var_real'].apply (lambda x :"Subiu" if x > 0 else ("Desceu" if x < 0 else "Estável"))
df_Principal
```

```python
df_Principal = df_Principal.merge(df_Ticker, left_on='Ativo',right_on='Ticker', how='left')
df_Principal = df_Principal.drop(columns=['Ticker'])
df_Principal
```

```python
df_Principal = df_Principal.merge(df_CHATGPT, left_on='Nome',right_on='Empresa', how='left')
df_Principal = df_Principal.drop(columns=['Empresa'])
df_Principal
```

```python
df_Principal = df_Principal.rename(columns={'Idade(em anos)':'Idade'}).copy()
df_Principal
```

```python
df_Principal['Cat_idade'] = pd.cut(df_Principal['Idade (em anos)'], bins=[-float('inf'), 50, 100, float('inf')], labels=['Menos de 50', 'Entre 50 e 100', 'Mais de 100 anos'])
df_Principal
```

```python
# Calculando o maior valor em var_real
maior = df_Principal['var_real'].max()

# Calculando o menor valor em var_real
menor = df_Principal['var_real'].min()

# Calculando a média de var_real
media = df_Principal['var_real'].mean()

# Calculando a média de var_real para os casos em que Resultado é 'Subiu'
media_subiu = df_Principal[df_Principal['Resultado'] == 'Subiu']['var_real'].mean()

# Calculando a média de var_real para os casos em que Resultado é 'Desceu'
media_desceu = df_Principal[df_Principal['Resultado'] == 'Desceu']['var_real'].mean()

# Imprimindo os resultados
print("Maior:", f"R$ {maior:,.2f}")
print("Menor:", f"R$ {menor:,.2f}")
print("Média:", f"R$ {media:,.2f}")
print("Média de quem subiu:", f"R$ {media_subiu:,.2f}")
print("Média de quem desceu:", f"R$ {media_desceu:,.2f}")
```

```python
#resultado de quem subiu
df_Principal_subiu = df_Principal[df_Principal['Resultado'] == 'Subiu']
df_Principal_subiu
```

```python
#variacao(R$) do segmento que subiu
df_analise_segmento = df_Principal_subiu.groupby('Segmento')['var_real'].sum().reset_index()
df_analise_segmento
```

```python
#variacao(R$) do saldo final
df_analise_saldo = df_Principal.groupby('Resultado')['var_real'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='Resultado', y='var_real', title='Variação(R$) por Resultado')
# Personalizando o formato dos números no eixo y
fig.update_layout(yaxis=dict(tickformat=",.2f"))  # Utiliza a formatação de milhares com 0 casas decimais
fig.update_layout(xaxis=dict(tickformat=",.2f"))  # Utiliza a formatação de milhares com 0 casas decimais
fig.show()
```

```python
fig = px.pie(df_analise_segmento, names='Segmento', values='var_real', title='Variação Reais por Segmento')
fig.show()
```

```python
df_analise_cat_idade = df_Principal.groupby('Cat_idade')['var_real'].sum().reset_index()
df_analise_cat_idade
```

```python
fig = px.bar(df_analise_cat_idade, x='Cat_idade', y='var_real', text='var_real', title='Variação Reais por Categoria de Idade')
fig.show()
```
